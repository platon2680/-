```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df.dropna(inplace = True)
# Обработка прпопущенных значений
```

```python
df['Genre'].value_counts()
df['Developer'].value_counts()
df['Publisher'].value_counts()
df['Series'].value_counts()
```

```python
# Проанализируем статью продаж
print('Maximum of Sales : ', df['Sales'].max())
print('Minimum of Sales : ', df['Sales'].min())
print('Median of Sales :', df['Sales'].median())
print('Average of Sales : ', df['Sales'].mean())
print('Total Sales : ', df['Sales'].sum())
```

```python
df['Sales'].hist(color = 'red')
# Вывод: Самое высокое распределение продаж от 0 до 5
```

```python
# Визуализация жанров игр на основе средних продаж
plt.figure(figsize = (13,5))
df.groupby('Genre')['Sales'].mean().sort_values(ascending = False).plot(kind = 'bar', color = 'orange')
plt.title('Визуализация жанров игр на основе средних продаж', fontsize = 15, color = 'red', pad = 12)
plt.xlabel('Жанр', fontsize = 12)
plt.ylabel('Средние продажи', fontsize = 12)
plt.show()
# Вывод: Самые распространенные жанры это: песочница, выживание. Далее идут симулятор жизни и ММО/РПГ.
```

```python
# Визуализация издателей игр на основе средних продаж
plt.figure(figsize = (15,5))
df.groupby('Publisher')['Sales'].mean().sort_values(ascending = False).plot(kind = 'bar', color = 'lime')
plt.title('Визуализация издателей игр на основе средних продаж', fontsize = 15, color = 'b', pad = 12)
plt.xlabel('Издатель', fontsize = 12)
plt.ylabel('Средние продажи', fontsize = 12)
plt.show()
# Самые крупные и имеющие наибольший успех издатели: Mojang, Atari, Windows и Blizzard.
```

```python
# Визуализация разработчиков игр на основе средних продаж
plt.figure(figsize = (15,5))
df.groupby('Developer')['Sales'].mean().sort_values(ascending = False).plot(kind = 'bar', color = 'orangered')
plt.title('Визуализация разработчиков игр на основе средних продаж', fontsize = 15, color = 'b', pad = 12)
plt.xlabel('Разработчик', fontsize = 12)
plt.ylabel('Средние продажи', fontsize = 12)
plt.show()
# Лучшими разработчиками игр (на основе средних продаж) считаются: Mojang, Frontier, Blizzard.
```

```python
# Визуализация серии игр на основе средних продаж
plt.figure(figsize = (18,5))
df.groupby('Series')['Sales'].mean().sort_values(ascending = False).plot(kind = 'bar', color = 'violet')
plt.title('Визуализация серии игр на основе средних продаж', fontsize = 15, color = 'r', pad = 12)
plt.xlabel('Серия', fontsize = 12)
plt.ylabel('Средние продажи', fontsize = 12)
plt.show()
# Лучшими сериями игр явялеются: Minecraft, Half-Life, Diablo и Симс.
```
